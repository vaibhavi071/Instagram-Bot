```python
# All the imports used

from selenium import webdriver
from getpass import getpass
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
driver = webdriver.Chrome(ChromeDriverManager().install())
import time
```

## Step 1: Login to your Instagram Handle
### Submit with sample username and password

```python
# Intialising Wait
wait=WebDriverWait(driver,10)
driver.get("https://www.instagram.com/")
text_fields = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"_2hvTZ")))
# Username and Password Fields
username = text_fields[0]
password = text_fields[1]
username.clear()
username.send_keys("SAMPLE_USERNAME")
password.clear()

# I have Implemented GETPASS, So password is also added anonymously as you can see in the output window
password.send_keys(getpass("Enter Password: "))
# You may also use this 
# password.send_keys("SAMPLE_PASSWORD")
login = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"sqdOP")))[1]
login.click()
not_now = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'cmbtv')]/button")))
not_now.click()
noti_not_now = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"HoLwm")))
noti_not_now.click() 
```

## Step 2: Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”

### Note : Make sure to avoid printing hashtags

```python
# Search Button
search=wait.until(EC.presence_of_element_located((By.CLASS_NAME,"x3qfX")))
search.clear()
search.send_keys("food")
usernames_a = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"yCE8d")))
# List of usernames
for ele in usernames_a:
    link=ele.get_attribute('href')
    if "tags" not in link:     # Removing TAGS
        print(link.split("/")[-2])
time.sleep(2)

search_close = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"aIYm8.coreSpriteSearchClear")))
search_close.click()
```

## Step 3:
### Searching and Opening a profile using 
### Open profile of “So Delhi” 

```python
# Function Created
def search_open(user):
    search=wait.until(EC.presence_of_element_located((By.CLASS_NAME,"x3qfX")))
    search.clear()
    search.send_keys(user)
    # Clicking on the result
    usernames_single = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"yCE8d")))[0]
    usernames_single.click()
```

# Step 5: Like/Unlike posts
### Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
### Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
# Searching and opening
search_open("dilsefoodie")
```

```python
# Liking
post=driver.find_elements_by_class_name("_9AhH0")
post[0].click()
for i in range(0,30):       
    # Getting Like Button
    like=wait.until(EC.presence_of_element_located((By.XPATH,'//section[contains(@class,"ltpMr Slqrh")]/span/button')))
    # Checking if alreayd liked or not usinf SVG Text
    svg= driver.find_element_by_css_selector(".eo2As > section > span > button > div > span > svg")
    if svg.get_attribute("aria-label")=='Unlike':
        print("Post " + str(i+1) + " is already liked")
    else:
        print("Post",i+1," liked")
        like.click()
    next_button=wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_65Bje.coreSpriteRightPaginationArrow")))
    next_button.click()
             
# Close button Click one time in the end
last_post_close = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'yiMZG')]/button")))
last_post_close.click()

    
```

```python
# Unliking
post=driver.find_elements_by_class_name("_9AhH0")
post[0].click()
for i in range(0,30):       
    # Getting Like Button
    
    like=wait.until(EC.presence_of_element_located((By.XPATH,'//section[contains(@class,"ltpMr Slqrh")]/span/button')))
    # Checking if alreayd liked or not usinf SVG Text
    svg= driver.find_element_by_css_selector(".eo2As > section > span > button > div > span > svg")
    if svg.get_attribute("aria-label")=='Like':
        print("Post " + str(i+1) + " is already unliked")
    else:
        print("Post",i+1," unliked")
        like.click()
    next_button=wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_65Bje.coreSpriteRightPaginationArrow")))
    next_button.click()
             
# Close button Click one time in the end
last_post_close = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'yiMZG')]/button")))
last_post_close.click()
```

# Step 6: Extract list of followers
### Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
### Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def follower_list(user):
    wait=WebDriverWait(driver,10)
    # Searching and opening
    driver.get("https://www.instagram.com/"+user+"/")
    follower_button=wait.until(EC.presence_of_element_located((By.XPATH,'//ul[contains(@class,"k9GMp ")]/li[2]/a')))
    follower_button.click()
    time.sleep(3)
    username_a = driver.find_elements_by_class_name("FPmhX.notranslate._0imsa")
    while len(username_a)<500:
        scroll2=driver.execute_script('return document.querySelector(".isgrP").scrollHeight;')
        scrollto2 = driver.execute_script('document.querySelector(".isgrP").scrollTo(0,arguments[0]);',scroll2)
        # time.sleep(3)          USE IF NET CONNECTION IS SLOW 
        username_a = driver.find_elements_by_class_name("FPmhX.notranslate._0imsa")

    username_a = driver.find_elements_by_class_name("FPmhX.notranslate._0imsa")[:500]
    i=0
    for ele in username_a:
        print(i+1,end="\t")
        link=ele.get_attribute('href')
        print(link.split("/")[-2])
        i+=1

    # Closing the followers window
    close_button = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'WaOAr')]/button")))
    close_button.click()
```

# Step 7: Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
### If You have already seen the story.
### Or The user has no story.
### Or View the story if not yet seen.

```python
search_open("coding.ninjas")
story_check =  wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'XjzKX')]/div")))
if story_check.get_attribute("class") == 'RR-M- ':
    print("The user has no story")
else:
    story=wait.until(EC.presence_of_element_located((By.CLASS_NAME,"CfWVH")))
    if story.get_attribute("height")=="210":
        print("Not seen the story")
        print("Started watching the stories of opened profile")
        i=0
        while story.get_attribute("height")=="210":
            story_check =  wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'XjzKX')]/div")))
            story_check.click()
            close_cross = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "coreSpriteCloseLight")))
            close_cross.click()
            print("Story " + str(i+1) + " seen")
            i=i+1
            story=wait.until(EC.presence_of_element_located((By.CLASS_NAME,"CfWVH")))
        print("All Stories Seen")



    elif story.get_attribute("height")=="208":
        print("Story seen")
```
